## Data Augmentation

```python

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range= 0.2,
                                   zoom_range= 0.2,
                                  horizontal_flip= True)
 
## in training data we reqired synthetic data to train the model well. 
## But in validation ot testing purpose we don't required synthetic data

val_datagen = ImageDataGenerator(rescale = 1./255)

test_datagen = ImageDataGenerator(rescale = 1./255)

## it follow the path and pick '32' images for every time training (batch_size)
## because it cant take all images togather
## then o/p the images as standard size = 224,224
## (class_mode) = it is a classification problem (COVID or NORMAL)

training_set = train_datagen.flow_from_directory('Dataset/Train',
                                                 target_size = (128,128),
                                                 batch_size = 32,
                                                 class_mode= 'categorical')
## also can be done as 128x128
val_set = train_datagen.flow_from_directory('Dataset/Val',
                                                 target_size = (128,128),
                                                 batch_size = 32,
                                                 class_mode= 'categorical')

test_set = train_datagen.flow_from_directory('Dataset/Test',
                                                 target_size = (128,128),
                                                 batch_size = 32,
                                                 class_mode= 'categorical')
```

```python
## split data = 80% for training, 10% validation, 10% testing
```

## Model Building

```python

from tensorflow.keras.applications.vgg16 import VGG16
```

```python

base_model = VGG16(input_shape= (128,128,3), weights= "imagenet",include_top = False)   
## 224 -width, 224 - height, 3 - RGB channel
## weights are from imagenet trained model
## include_top = False (means last layer is excluded)
```

```python

for layer in base_model.layers:
     layer.trainable = False   
## every hidden layer contain parameter trainable.
## By using loop we can FREEZE this parameter in each layer 
```

```python

inp = base_model.input
## in base_model we just pass our images 

x = base_model.output
## x is the output of the base_model

x = Flatten()(x)
## Flatten convert the 3x3 matrix in a single dimention vector 

x = Dense(1024, activation = 'relu')(x)
## after Flatten output x, it contains all the features.
## dense mapped those features to the classes

x = Dropout(0.5)(x)
## Dropout layer: it eliminate some neurons at every iteration randomly to adjust OVERFITTING
## here 0.5 means 50% neurons are eliminated

output = Dense(2, activation = 'softmax')(x) ## 2 because only required 2 neurons
## 2 neurons bcoz we have 2 classes only

model = Model(inputs = inp, outputs = output)
```

```python
model.summary()
```

```python

model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)
## categorical_crossentropy : loss function
## optimizer : speed the training
```

```python

history = model.fit(
    training_set,
    batch_size = 32,
    verbose = 1,
    epochs = 5,
    validation_data = (val_set)
)
## verbose (values - 0,1,2): 0 - silent, 1 - show progress, 2 - more details
```

```python
## (3641/32) = 113.78 = 114
```

```python
history.history
```

```python

print("Evaluate on test data ")
results = model.evaluate(test_set, batch_size=32)
print("test loss, test accuracy : ",results)
```

```python

model.save('model.h5')
```

```python

def plot(history):
    fig, axes = plt.subplots(1, 2, figsize = (10, 4))
    axes[0].plot(history["loss"], "r-", history["val_loss"], "b-", linewidth = 0.5)
    axes[0].set_title("Loss Curve")
    axes[1].plot(history["accuracy"], "r-", history["val_accuracy"], "b-", linewidth = 0.5)
    axes[1].set_title("Accuracy Curve")
    fig.tight_layout()
    fig.show()
```

## ROC - AUC

```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import roc_curve, auc, roc_auc_score
```

```python
target = ['ROC curve']

##set plot figure size
fig, c_ax = plt.subplots(1,1,figsize = (12,8))

##function for scoring roc auc score for multi-class
def roc(y_test , y_pred, average = 'macro'):
    lb = LabelBinarizer()
    lb.fit(y_test)
    y_test = lb.transform(y_test)
    y_pred = lb.transform(y_pred)
    
    for(idx, c_label) in enumerate(target):
        fpr, tpr, thresholds = roc_curve(y_test[:,idx].astype(int), y_pred[:,idx])
        c_ax.plot(fpr, tpr, label = '%s (AUC:%0.2f)' % (c_label, auc(fpr, tpr)))

    c_ax.plot(fpr, fpr, 'b-', label = 'Random Guessing')
    return roc_auc_score(y_test, y_pred, average= average)

print('ROC AUC Score : ', roc(y_test,y_pred))

c_ax.legend()
c_ax.set_xlabel('False Positive Rate')
c_ax.set_ylabel('True Positive Rate')
plt.show()
```
